```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

```python
last_fact = 1
cache = {last_fact: 1}

def fact(N):
    global last_fact, cache
    if N < 2:
        return 1
    if N in cache:
        return cache[N]
    result = cache[last_fact]
    for i in range(last_fact+1, N+1):
        result = result * i
        cache[i] = result
    return cache[N]

def f(N):
    result = fact(N)
    nb0s = 0
    while result % 10 == 0:
        result = result // 10
        nb0s += 1
    return nb0s
```

```python
x2 = [x*x for x in range(100)]
plt.plot(x2)
plt.ylabel('x*x')
plt.show()
```

```python
fx = [f(x) for x in range(10000)]
plt.plot(fx)
plt.ylabel('trailing_0s')
plt.show()
```
